# Practica Hospitales

A partir del CSV hospitales2.csv, generar un archivo CSV de salida, que contenga los siguientes campos en este orden:
* latitude
* longitude
* name
* label

Con los correspondientes datos extraídos del CSV original, el campo name tiene que corresponder con la dirección del hospital, y el campo label con el nombre del hospital.

---

```python
import pandas as pd

df = pd.read_csv("hospitales2.csv")

df.head()
```

---
#### Recuperamos cantidad de registros, cantidad de columnas, nombre de las columnas

---

```python
n_rows, n_cols = df.shape
cols = list(df.columns)
sep = "-------------------------------------------------------------------------------------"
print(sep)
print("Cantidad de columnas (features) --> ", n_cols)
print(sep)
print("Cantidad de filas --> ", n_rows)
print(sep)
print("Nombre de las columnas --> \n", cols)
print(sep)
```

---
#### La informacion que necesitamos se encuentra en algunas columnas

Necesitamos las siguientes columnas:
* WKT
* NOMBRE
* CALLE
* ALTURA

---

#### Generamos un nuevo dataframe, solo con las columnas necesarias

---

```python
df2 = df[["WKT", "NOMBRE", "CALLE", "ALTURA"]]
df2.head()
```

---
#### Genero dos funciones para recuperar la latitud y la longitud

---

```python
str = "POINT (-58.3775508488443 -34.6288473603881)"




def getLatitud(str):
    ini = str.rfind("(")+1
    fin = str.find(")")
    lista = str[ini:fin].split(" ")
    latitud = float(lista[0])
    

    return latitud

def getLongitud(str):
    ini = str.rfind("(")+1
    fin = str.find(")")
    lista = str[ini:fin].split(" ")    
    longitud = float(lista[1])
    return longitud


getLatitud(str)


```

---
#### Creamos dos nuevas columnas latitud y longitud, a partir de la columna WKT utilizando las funciones previamente creadas


---

```python
df2["latitud"] = df2["WKT"].apply(getLatitud)
df2["longitud"] = df2["WKT"].apply(getLongitud)
```

```python
df2.head()
```

---
#### Creo una nueva columna con la direccion entera (calle + altura)


---

```python
df2["Name"] = df2["CALLE"]+" "+df2["ALTURA"].astype("str")
```

```python
df2.head()
```

---
#### Renombro la coluna NOMBRE a label (es lo solicitado en el enunciado)

---

```python
df2.rename(columns={"NOMBRE":"label"}, inplace=True)
df2.head()
```

---
#### Elimino las columnas que no necesito

---

```python
df2 = df2.drop(["WKT", "CALLE", "ALTURA"], axis=1)
df2.head()
```

---
#### Reordeno las columnas para que esten en el orden solicitado

---

```python
df2 = df2[["latitud", "longitud", "label", "Name"]]
df2.head()
```

---
#### Finalmente guardo el dataframe en un archivo .csv (separado por comas)

---

```python
df2.to_csv("hospital_procesado_noe.csv", sep=",", index=False)
```
