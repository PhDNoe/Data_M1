1) Implementar un juego, que consista en apilar números enteros del 1 al 20, de forma aleatoria, para lo cual debe usarse una estructura de Pila. 
Luego, el usuario debe elegir un número de veces en que se va a quitar elementos de la pila, los cuales, sumados entre sí, no deben superar el valor de 50.
El usuario pierde si la suma supera ese valor. Si no lo supera, gana, pero su calificación será 10 menos el número elementos que falten quitar para todavía no superar 50.
El programa debe informar si perdió, y si ganó, con qué calificación lo hizo.

Consideraciones:<br>
a. Se puede usar la función input() para obtener una entrada de teclado.<br>
b. Se puede usar la el modulo random para obtener valores aleatorios.

```python
import random

class Juego():

    def __init__(self):
        self.__lista = []
        
    # Pila de numeros del 1 al 20
    def creo_pila(self):
        numeros = list(range(1,21))
        random.shuffle(numeros)
        self.lista_orig = numeros.copy()
        return numeros

    # Muestro la lista original de los numeros
    def show_numeros(self):
        print("Lista de numeros originales --> ",self.lista_orig)

    # Desapilo n numeros y los sumo
    def pop_y_suma(self, n):
        
        suma=0
        self.__lista = []
        for i in range(n):
            seleccion = self.numeros.pop()
            suma += seleccion
            self.__lista.append(seleccion)

        return suma

    # Muestro numeros desapilados
    def show_lista(self, n):
        print(f"Lista de los {n} numeros desapilados")
        print(self.__lista)

    def puntaje(self):

        puntos = 10
        suma_full = self.suma

        if self.suma>50:
            return -1
        else:
            i=0
            while suma_full<50:
                suma_full += self.numeros.pop()
                i += 1
        return puntos - i

    # Juego
    def jugar(self):

        

        n = int(input("Por favor ingrese la cantidad de numeros que debemos sacar de la pila"))

        self.numeros = self.creo_pila()

        self.suma = self.pop_y_suma(n)

        if self.suma>50:
            print(f"Ha Perdido el juego!!! Ha sumado {self.suma} puntos")
        else:
            self.puntos = self.puntaje()
            print(f"Ha ganado. Ha sumado {self.suma} puntos - Su score es {self.puntos}")
        self.show_numeros()
        self.show_lista(n)
        

```

```python
juego = Juego()

juego.jugar()
```

```python
sep = "--------------------------------------------------------------------------------"
# Pasos 1 y 2 - Vaciamos las jarras
Jarra_3L = []
Jarra_5L = []

max_size_5L = 5
max_size_3L = 3

actual_content_5L = len(Jarra_5L)
actual_content_3L = len(Jarra_3L)

# 3 - Llenamos la Jarra de 3L
while (actual_content_3L<max_size_3L):
    Jarra_3L.append(1)
    actual_content_3L = len(Jarra_3L)

print(sep)
print("3)")
print("Jarra_3L --> ", Jarra_3L)
print("Jarra_5L --> ", Jarra_5L)

# 4 - Vertemos la Jarra de 3L en la de 5L
while (actual_content_3L>0 and actual_content_5L<max_size_5L):
    litro = Jarra_3L.pop()
    Jarra_5L.append(litro)
    actual_content_5L = len(Jarra_5L)
    actual_content_3L = len(Jarra_3L)

print(sep)
print("4)")
print("Jarra_3L --> ", Jarra_3L)
print("Jarra_5L --> ", Jarra_5L)

# 5 - Llenamos la Jarra de 3L
while (actual_content_3L<max_size_3L):
    Jarra_3L.append(1)
    actual_content_3L = len(Jarra_3L)

print(sep)
print("5)")
print("Jarra_3L --> ", Jarra_3L)
print("Jarra_5L --> ", Jarra_5L)

# 6 - Vertemos la Jarra de 3L en la de 5L
while (actual_content_3L>0 and actual_content_5L<max_size_5L):
    litro = Jarra_3L.pop()
    Jarra_5L.append(litro)
    actual_content_5L = len(Jarra_5L)
    actual_content_3L = len(Jarra_3L)

print(sep)
print("6)")
print("Jarra_3L --> ", Jarra_3L)
print("Jarra_5L --> ", Jarra_5L)

# 7 - Vaciar la Jarra de 5L
Jarra_5L = []
actual_content_5L = len(Jarra_5L)

print(sep)
print("7)")
print("Jarra_3L --> ", Jarra_3L)
print("Jarra_5L --> ", Jarra_5L)

# 8 - Vertemos la Jarra de 3L en la de 5L
while (actual_content_3L>0 and actual_content_5L<max_size_5L):
    litro = Jarra_3L.pop()
    Jarra_5L.append(litro)
    actual_content_5L = len(Jarra_5L)
    actual_content_3L = len(Jarra_3L)

print(sep)
print("8)")
print("Jarra_3L --> ", Jarra_3L)
print("Jarra_5L --> ", Jarra_5L)


# 9 - Llenamos la Jarra de 3L
while (actual_content_3L<max_size_3L):
    Jarra_3L.append(1)
    actual_content_3L = len(Jarra_3L)

print(sep)
print("9)")
print("Jarra_3L --> ", Jarra_3L)
print("Jarra_5L --> ", Jarra_5L)

# 10 - Vertemos la Jarra de 3L en la de 5L
while (actual_content_3L>0 and actual_content_5L<max_size_5L):
    litro = Jarra_3L.pop()
    Jarra_5L.append(litro)
    actual_content_5L = len(Jarra_5L)
    actual_content_3L = len(Jarra_3L)

print(sep)
print("10)")
print("Jarra_3L --> ", Jarra_3L)
print("Jarra_5L --> ", Jarra_5L)
print("Cantidad de Litros en la Jarra 5 --> ", len(Jarra_5L))
```

```python
class juego_jarras():

    def __init__(self):
        self.Jarra_3L = []
        self.Jarra_5L = []
        self.max_size_3L = 3
        self.max_size_5L = 5
        self.actual_content_3L = 0
        self.actual_content_5L = 0
        self.sep = "-------------------------------------------------------------------"

    def muestra_contenido(self):
        
        print(self.sep)
        print("Jarra de 3L --> ", self.Jarra_3L)
        print("Jarra de 5L --> ", self.Jarra_5L)
        print(self.sep)

    def gano_juego(self):
        if self.actual_content_5L ==4:
            print("Felicitaciones!!! Ha logrado colocar 4L en una de las Botellas")


    def vaciar_jarra_5L(self):
        self.Jarra_5L = []
        self.actual_content_5L = 0
        self.muestra_contenido()

    def vaciar_jarra_3L(self):
        self.Jarra_3L = []
        self.actual_content_3L = 0
        self.muestra_contenido()

    def llenar_jarra_3L(self):
        while (self.actual_content_3L<self.max_size_3L):
            self.Jarra_3L.append(1)
            self.actual_content_3L = len(self.Jarra_3L)
        self.muestra_contenido()

    def llenar_jarra_5L(self):
        while (self.actual_content_5L<self.max_size_5L):
            self.Jarra_5L.append(1)
            self.actual_content_5L = len(self.Jarra_5L)
        self.muestra_contenido()

    def verter_3L_en_5L(self):
        
        while (self.actual_content_3L>0 and self.actual_content_5L<self.max_size_5L):
            litro = self.Jarra_3L.pop()
            self.Jarra_5L.append(litro)
            self.actual_content_5L = len(self.Jarra_5L)
            self.actual_content_3L = len(self.Jarra_3L)

        self.muestra_contenido()
        self.gano_juego()

    def verter_5L_en_3L(self):
        
        while (self.actual_content_5L>0 and self.actual_content_3L<self.max_size_3L):
            litro = self.Jarra_5L.pop()
            self.Jarra_3L.append(litro)
            self.actual_content_5L = len(self.Jarra_5L)
            self.actual_content_3L = len(self.Jarra_3L)
        self.muestra_contenido()
```
